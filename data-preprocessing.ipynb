Filtering values out of dataset that are not continuous. Also changed the target values to 1 and 0 for graduated/enrolled and dropout respectively.

````python
import numpy as np
import pandas as pd
````

````python
# Importing raw data set of student dropout and academic successe
raw_data = pd.read_csv('data/data.csv')
print(raw_data)
````

````python
# Removing non-ordinal (nominal features)
nonordinal_features = ['Displaced', 'Educational special needs', 'Debtor', 'Tuition fees up to date', 'Gender', 
                    'Scholarship holder', 'International', 'Marital Status', 'Daytime/evening attendance	', 'Application mode', 'Application order',
                    'Course', 'Previous qualification', "Mother's qualification", "Father's qualification",
                    "Mother's occupation", "Father's occupation", 'Nacionality', 'Marital status']

filtered_dataset = raw_data.drop(columns=nonordinal_features, errors='ignore', axis=1)
print(filtered_dataset)
````

````python
filtered_dataset['Target'] = filtered_dataset['Target'].map({'Graduate': 1, 'Enrolled': 1, 'Dropout': 0})
````

Renaming labels to 1 for graduated/enrolled and 0 for dropout. Checking percent of 1 labels.

````python

labels = filtered_dataset['Target']

count = 0
for index, value in labels.items():
    if value == 1:
        count+= 1

print(f'percent of graduate/enrolled students is {count/len(labels)}')
````

Now, the data gets normalized, 

````python
from sklearn.preprocessing import MinMaxScaler

features = filtered_dataset.columns.drop('Target')

# Normalize scale of columns (so that all values are scaled between 0 and 1). Labels stay attached so that data can be split up.
Scaler = MinMaxScaler()
filtered_dataset[features] = Scaler.fit_transform(filtered_dataset[features])


filtered_dataset.head()
````

Seperating the training and test sets, then performing 5 fold cross validation.
Since each fold is a np array, currently accessing the split needs to be done by index. 0 represents the training split, 1 represents the test split.

For example - to access the training data in the 3rd fold, call this: folds[2][0]. The first index is the fold, the second is the split.

````python
from sklearn.model_selection import KFold

data = np.array(filtered_dataset) #convert to numpy array
m = int(0.8 * len(filtered_dataset)) #80/20 split
training_data = data[:m]
test_data = data[m:]


# KFold split
kf = KFold(n_splits=5)
folds = []
for i, (train_index, val_index) in enumerate(kf.split(training_data)):
    train_split, val_split = training_data[train_index], training_data[val_index]
    folds.append((train_split, val_split))
    print(f"Fold {i+1}:\nTrain shape: {train_split.shape}, Val shape: {val_split.shape}")
    print(f"Train indices: {train_index}")
    print(f"Validation indices: {val_index[:10]} ... {val_index[-10:]}\n")
````
